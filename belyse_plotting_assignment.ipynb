```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/LaurelleJinelle/Project-Plotting---Matplotlib-Seaborn-Histograms-and-Boxplots/refs/heads/main/belyse_fuel_econ.csv')
df.head(11)
```

```python
plt.hist(df['co2'], bins=20)
plt.xlabel('CO2')
plt.ylabel('Frequency')
plt.title('Histogram for CO2', fontsize=24)
plt.show
```

```python
plt.hist(df['cylinders'], bins=6)
plt.xlabel('Cylinder')
plt.ylabel('Frequency')
plt.title('Histogram For Cylinders',fontsize=24)
plt.show
```

```python
df_numeric = df.select_dtypes(include=np.number)
values = df_numeric.corr()
sns.heatmap(values)
plt.show()
```
