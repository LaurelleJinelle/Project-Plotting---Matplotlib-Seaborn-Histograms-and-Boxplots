```python
# YOUR CODE HERE
plt.hist(df['co2'], bins=20)
plt.xlabel('CO2')
plt.ylabel('Frequency')
plt.title('Histogram For CO2')
plt.show
```

```python
# YOUR CODE HERE
plt.hist(df['make'], bins=10)
plt.xlabel('Make of Vehicle')
plt.ylabel('Frequency')
plt.title('Histogram For Make Of Vehicles')
plt.show
```

```python
# YOUR CODE HERE
plt.hist(df['cylinders'], bins=6)
plt.xlabel('Cylinder')
plt.ylabel('Frequency')
plt.title('Histogram For Cylinders')
plt.show
```
